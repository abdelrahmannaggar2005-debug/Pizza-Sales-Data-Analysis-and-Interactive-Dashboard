```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('pizza_sales.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df = df.drop('Unnamed: 13',axis =1)
df = df.drop('Unnamed: 14',axis =1)
df.isna().sum()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['order_time'] = pd.to_datetime(df['order_time'])
df['day_of_week'] = df['order_date'].dt.day_name()
df['hour'] = df['order_time'].dt.hour
df
```

```python
total_revenue = df['total_price'].sum()
total_orders = df['order_id'].nunique()
avg_order_value = total_revenue / total_orders
total_pizzas = df['quantity'].sum()
avg_pizzas_per_order = total_pizzas / total_orders

print(f"Total Revenue: {total_revenue:.2f}")
print(f"Average Order Value: {avg_order_value:.2f}")
print(f"Total Pizzas Sold: {total_pizzas}")
print(f"Total Orders: {total_orders}")
print(f"Average Pizzas Per Order: {avg_pizzas_per_order:.2f}")
```

```python
daily_orders = df.groupby('day_of_week')['order_id'].nunique()
daily_orders
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=daily_orders.index, y=daily_orders.values, palette='viridis')
plt.title('Total Orders by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Total Orders')
plt.show()
```

```python
hourly_orders = df.groupby('hour')['order_id'].nunique()
hourly_orders
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=hourly_orders.index, y=hourly_orders.values, marker='o')
plt.title('Total Orders by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Total Orders')
plt.xticks(range(0, 24))
plt.show()
```

```python
category_sales = df.groupby('pizza_category')['total_price'].sum()
category_pct = (category_sales / total_revenue) * 100
category_pct
```

```python
plt.figure(figsize=(8, 8))
category_pct.plot(kind='pie', autopct='%1.2f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('% of Sales by Pizza Category')
plt.ylabel('')
plt.show()
```

```python
pizza_sold = df.groupby('pizza_name')['quantity'].sum()
pizza_sold
```

```python
top_5 = pizza_sold.sort_values(ascending=False).head(5)
bottom_5 = pizza_sold.sort_values(ascending=True).head(5)

print("Top 5 Pizzas:")
print(top_5)
print("Bottom 5 Pizzas:")
print(bottom_5)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5.values, y=top_5.index, palette='Blues_d')
plt.title('Top 5 Best Selling Pizzas')
plt.xlabel('Total Sold')
plt.ylabel('Pizza Name')
plt.show()
```
